```python
import pandas as pd

# Charger les donn√©es dans un dataframe
data = pd.read_csv('video-games.csv') 
```

```python
# Clean dataset
data.dropna (subset=['Global_Sales'], inplace=True)
```

```python
# Area chart to show trend global sales over the years 
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.fill_between(data['Year'], data['Global_Sales'], color='skyblue', alpha=0.6)
plt.plot(data['Year'], data['Global_Sales'], color='Slateblue', alpha=0.8)
plt.title('Trend of Global Sales Over the Years')
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.show()
```

```python
# histogram to show the distribution of global sales
plt.figure(figsize=(8, 5))
plt.hist(data['Global_Sales'], bins=20, color='lightcoral')
plt.title('Distribution of global saleses')
plt.xlabel('global saless')
plt.ylabel('Frequency')
plt.show()
```

```python
# scatter plot to show the relationship between North American sales and European sales
plt.figure(figsize=(8, 6))
plt.scatter(data['NA_Sales'], data['EU_Sales'], color='mediumseagreen')
plt.title('Relationship between North American sales and European sales')
plt.xlabel('North American sales')
plt.ylabel('European sales')
plt.show()
```

```python
# scatter plot to show the relationship between Japanese sales and Other sales
plt.figure(figsize=(8, 6))
plt.scatter(data['JP_Sales'], data['Other_Sales'], color='gold')
plt.title('Relationship between Japanese sales and Other sales')
plt.xlabel('Japanese sales')
plt.ylabel('Other saless')
plt.show()

```

```python
#stacked area chart to show the distribution of sales by platform over the years

import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Global_Sales', hue='Platform', data=data.groupby(['Year', 'Platform'], as_index=False).sum(), palette='Set2')
plt.title('Distribution of sales by platform over the years')
plt.xlabel('Year')
plt.ylabel('Global-Sales')
plt.legend(title='Platform')
plt.show()
```

```python
#bar chart to show the top 10 publishers by global sales

top_publishers = data.groupby('Publisher')['Global_Sales'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
top_publishers.plot(kind='bar', color='royalblue')
plt.title('Top 10 publishers by global sales')
plt.xlabel('Publisher')
plt.ylabel('Global Sales')
plt.show()
```

```python
#pie chart to show the distribution of genres by global sales
genre_sales = data.groupby('Genre')['Global_Sales'].sum()
plt.figure(figsize=(8, 8))
plt.pie(genre_sales, labels=genre_sales.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('distribution of genres by global sales')
plt.axis('equal')
plt.show()
```
